```python
## Importing Standard Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set_style('darkgrid')
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:/Users/INDIA/Downloads/Global Terrorism - START data/globalterrorismdb_0718dist.csv")
```

```python
df.head()
```

```python
## Transpose the data to read conveniently
df.describe().transpose()['max']
```

```python
## Statistical description of the dataset
df.describe()
```

```python
## Checking for NaN Values
df.isna().sum()
```

```python
## Removing NaN values
df=df.dropna(axis=1)
df.isna().sum()
```

```python
plt.figure(figsize=(40,10), dpi=300)
sns.histplot(df,x='country_txt')
plt.xticks(rotation=90)
plt.show()
```

**In oreder to get the readable and precise data we will sort according to number of terrorism occured in a country**

```python
group1=df.groupby('country_txt').count()
group1
```

```python
## Occurance of terror events in the country

country_count=df['country_txt'].value_counts()
country_count
```

```python
## Save this dataset as a new file
country_count.to_csv('file1.csv')
```

```python
## Import the new file
most_terror=pd.read_csv('file1.csv')
```

```python
most_terror.head()
```

```python
## Rename the columns
most_terror.rename(columns={'Unnamed: 0':'Country','country_txt':'Terror_count'},inplace=True)
```

```python
most_terror1=most_terror.head(10)
most_terror1
```

```python
## Find out the Hot zone terrorisms
plt.figure(figsize=(8,4),dpi=300)
plt.scatter(data=most_terror1,x='Country',y='Terror_count',marker='D')
plt.xticks(rotation=90)
plt.title('Top 10 Terrorism Hot Zones')
plt.xlabel('Country')
plt.ylabel('Number of terrorist activity')
plt.show()
```
